# 03. YOLO11 / RF-DETR 시각화 영상 좌우 비교

이 노트북은 이미 생성된 두 개의 시각화 영상을 좌우로 붙여 하나의 비교 영상을 만듭니다.

- 왼쪽 기본 제목: `YOLO11`
- 오른쪽 기본 제목: `RF-DETR`
- 위쪽 title bar에 모델 이름을 표시합니다.
- 두 영상 크기가 다르면 같은 높이로 resize한 뒤 합칩니다.

이 노트북은 model을 load하지 않으므로 YOLO/RF-DETR 전용 env가 아니어도 실행할 수 있습니다. OpenCV와 Pillow만 있으면 됩니다.


```python
from pathlib import Path
import sys


def find_repo_root(start: Path) -> Path:
    for path in [start, *start.parents]:
        if (path / "pyproject.toml").is_file() and (path / "src").is_dir():
            return path
    raise RuntimeError("repo root를 찾지 못했습니다. notebook을 repo 안에서 실행하거나 REPO_ROOT를 직접 지정하세요.")

REPO_ROOT = find_repo_root(Path.cwd())
src_path = str(REPO_ROOT / "src")
if src_path not in sys.path:
    sys.path.insert(0, src_path)

print("REPO_ROOT:", REPO_ROOT)

from labelstudio_bbox_tools.video_inference.compare import combine_videos_side_by_side
```

```python
# ===== 사용자가 주로 바꾸는 설정 =====

LEFT_VIDEO = Path("/path/to/yolo11_visualized.mp4")
RIGHT_VIDEO = Path("/path/to/rfdetr_visualized.mp4")
OUT_VIDEO = Path("/path/to/compare_yolo11_rfdetr.mp4")

LEFT_TITLE = "YOLO11"
RIGHT_TITLE = "RF-DETR"
FONT_PATH = None
FONT_SIZE = 28
TITLE_BAR_HEIGHT = 48
TARGET_HEIGHT = None   # None이면 두 영상 중 작은 height에 맞춥니다.
FPS = None             # None이면 왼쪽 영상 fps 우선 사용.
MAX_FRAMES = 300       # 전체 비교 영상이면 None.
OVERWRITE = False

RUN_PREVIEW = False
RUN_COMPARE = False
DRY_RUN = True
```

```python
# 두 영상 metadata와 출력 크기만 확인합니다. 영상은 쓰지 않습니다.

if not RUN_PREVIEW:
    print("RUN_PREVIEW=False 이므로 preview를 실행하지 않습니다.")
else:
    preview = combine_videos_side_by_side(
        left_video=LEFT_VIDEO,
        right_video=RIGHT_VIDEO,
        out_video=OUT_VIDEO,
        left_title=LEFT_TITLE,
        right_title=RIGHT_TITLE,
        font_path=FONT_PATH,
        font_size=FONT_SIZE,
        title_bar_height=TITLE_BAR_HEIGHT,
        target_height=TARGET_HEIGHT,
        fps=FPS,
        max_frames=MAX_FRAMES,
        dry_run=True,
    )
    print(preview.as_dict())
```

```python
# 실제 좌우 비교 영상 저장 cell입니다.

if not RUN_COMPARE:
    print("RUN_COMPARE=False 이므로 비교 영상을 생성하지 않습니다.")
else:
    summary = combine_videos_side_by_side(
        left_video=LEFT_VIDEO,
        right_video=RIGHT_VIDEO,
        out_video=OUT_VIDEO,
        left_title=LEFT_TITLE,
        right_title=RIGHT_TITLE,
        font_path=FONT_PATH,
        font_size=FONT_SIZE,
        title_bar_height=TITLE_BAR_HEIGHT,
        target_height=TARGET_HEIGHT,
        fps=FPS,
        max_frames=MAX_FRAMES,
        overwrite=OVERWRITE,
        dry_run=DRY_RUN,
    )
    print(summary.as_dict())
```
