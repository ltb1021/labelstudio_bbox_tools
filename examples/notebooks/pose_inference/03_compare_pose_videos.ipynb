# 03 Pose 시각화 영상 좌우 비교

이미 생성된 YOLO11 pose/RF-DETR Keypoint 결과 영상을 좌우로 합쳐 비교합니다. 모델을 load하지 않습니다.

```python
from pathlib import Path
import sys


def find_repo_root(start: Path) -> Path:
    for path in [start, *start.parents]:
        if (path / "pyproject.toml").is_file() and (path / "src").is_dir():
            return path
    raise RuntimeError("repo root를 찾지 못했습니다. notebook을 repo 안에서 실행하거나 REPO_ROOT를 직접 지정하세요.")


REPO_ROOT = find_repo_root(Path.cwd())
src_path = str(REPO_ROOT / "src")
if src_path not in sys.path:
    sys.path.insert(0, src_path)

print("REPO_ROOT:", REPO_ROOT)

from labelstudio_bbox_tools.video_inference.compare import combine_videos_side_by_side
```

## 설정

두 결과 영상 경로와 저장할 비교 영상 경로를 지정합니다.

```python
# YOLO11 pose notebook과 RF-DETR pose notebook에서 생성된 mp4 경로를 넣습니다.
LEFT_VIDEO = Path("/path/to/yolo11_pose_visualized.mp4")
RIGHT_VIDEO = Path("/path/to/rfdetr_keypoint_visualized.mp4")
OUT_VIDEO = Path("/path/to/compare_yolo11_rfdetr_pose.mp4")

LEFT_TITLE = "YOLO11 Pose"
RIGHT_TITLE = "RF-DETR Keypoint"

FONT_PATH = None
FONT_SIZE = 28
TITLE_BAR_HEIGHT = 48
TARGET_HEIGHT = None   # None이면 두 영상 중 더 낮은 높이에 맞춥니다.
FPS = None             # None이면 왼쪽 영상 fps 우선 사용
MAX_FRAMES = 300
OVERWRITE = False

RUN_COMPARE = False
DRY_RUN = True
```

## 비교 영상 생성

처음에는 `RUN_COMPARE=True`, `DRY_RUN=True`로 크기/fps/frame 수만 확인하고, 실제 저장할 때 `DRY_RUN=False`로 바꿉니다.

```python
if RUN_COMPARE:
    summary = combine_videos_side_by_side(
        left_video=LEFT_VIDEO,
        right_video=RIGHT_VIDEO,
        out_video=OUT_VIDEO,
        left_title=LEFT_TITLE,
        right_title=RIGHT_TITLE,
        font_path=FONT_PATH,
        font_size=FONT_SIZE,
        title_bar_height=TITLE_BAR_HEIGHT,
        target_height=TARGET_HEIGHT,
        fps=FPS,
        max_frames=MAX_FRAMES,
        overwrite=OVERWRITE,
        dry_run=DRY_RUN,
    )
    summary.as_dict()
else:
    print("RUN_COMPARE=False 이므로 비교 영상 생성을 건너뜁니다.")
```
